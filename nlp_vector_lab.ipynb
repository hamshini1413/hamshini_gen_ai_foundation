```python
import pandas as pd
from pathlib import Path
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

```python
csv_path = Path("Reviews.csv")
if not csv_path.exists():
    csv_path = Path("lab1/Reviews.csv")

df = pd.read_csv(csv_path, nrows=50000)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
light_reviews = df[df["Text"].str.contains("light", case=False, na=False)]

print("Number of reviews containing 'light':", len(light_reviews))

light_reviews[["Text"]].head(20)
```

```python
sentences = [

"This olive oil is wonderful and very light on the stomach.",

"The packaging was broken and let too much light inside the box.",

"I prefer this snack because it is light and healthy."

]
```

```python
print("Sentence 1:")
print(sentences[0])

print("\nSentence 2:")
print(sentences[1])

print("\nSentence 3:")
print(sentences[2])
```

```python
vectorizer = CountVectorizer()

document_term_matrix = vectorizer.fit_transform(sentences)

print("Vectorization Completed")
```

```python
dense_matrix = document_term_matrix.toarray()

print("Dense Matrix:\n")
print(dense_matrix)
```

```python
feature_names = vectorizer.get_feature_names_out()

print(feature_names)
```

```python
vec_1 = dense_matrix[0]

vec_2 = dense_matrix[1]

vec_3 = dense_matrix[2]

print("Vector 1:")
print(vec_1)

print()

print("Vector 2:")
print(vec_2)

print()

print("Vector 3:")
print(vec_3)
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)
```

```python
print("="*50)

print("Dot Product Similarity")

print("="*50)

print()

print("Sentence 1 vs Sentence 2 (Different Context):")

print(similarity_1_2)

print()

print("Sentence 1 vs Sentence 3 (Similar Context):")

print(similarity_1_3)
```

```python
matrix_df = pd.DataFrame(
    dense_matrix,
    columns=feature_names,
    index=["Sentence 1","Sentence 2","Sentence 3"]
)

matrix_df
```

```python
print("Conclusion")

if similarity_1_3 > similarity_1_2:
    print("Sentence 1 is mathematically more similar to Sentence 3.")
else:
    print("Sentence 1 is mathematically more similar to Sentence 2.")

print("\nBag-of-Words cannot understand the contextual meaning of words.")
print("It only counts word occurrences.")
print("Modern LLMs solve this problem using Context-Aware Embeddings and Self-Attention.")
```
