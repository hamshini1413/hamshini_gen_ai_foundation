```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python

df = pd.read_csv("Reviews.csv", nrows=50000)

print("Dataset Shape:", df.shape)
print(df.columns)
```

```python
# Filter reviews containing the word "light"
light_reviews = df[df["Text"].str.contains("light", case=False, na=False)]

print("Number of reviews containing 'light':", len(light_reviews))

# Display first 20 matching reviews
for i, review in enumerate(light_reviews["Text"].head(20)):
    print(f"\nReview {i+1}:")
    print(review)
```

```python
sentences = [
    "This olive oil is wonderful and very light on the stomach.",
    "The packaging was broken and let too much light inside the box.",
    "I prefer this brand because it is a light and healthy snack."
]

print("Selected Sentences:")
for s in sentences:
    print("-", s)
```

```python
vectorizer = CountVectorizer()

matrix = vectorizer.fit_transform(sentences)

dense_matrix = matrix.toarray()

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nDense Matrix:")
print(dense_matrix)
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("Vector 1:", vec_1)
print("Vector 2:", vec_2)
print("Vector 3:", vec_3)
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)

print("\nSimilarity Scores")
print("------------------")
print("Sentence 1 vs Sentence 2:", similarity_1_2)
print("Sentence 1 vs Sentence 3:", similarity_1_3)
```

```python
print("\nConclusion:")
print("CountVectorizer treats words as counts.")
print("It cannot understand that the word 'light' has different meanings in different contexts.")
print("This demonstrates the limitation of Bag-of-Words representations.")
```
